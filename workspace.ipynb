```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# training loss vs epoch results after running LSTM and QLSTM
df = pd.read_csv('LSTM_loss.csv')
classical_train_loss = df.iloc[:, 0].to_numpy()

df = pd.read_csv('QLSTM_loss.csv')
quantum_train_loss = df.iloc[:, 0].to_numpy()
```

```python
x = np.arange(1, len(classical_train_loss) + 1)

# Plot loss vs epoch
plt.plot(x, classical_train_loss, marker='o', label='Classical', color='red')
plt.plot(x, quantum_train_loss, marker='o', label='Quantum', color='indigo')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Classical and Quantum LSTM Training Loss vs Epoch')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plot log loss vs epoch
plt.plot(x, np.log10(classical_train_loss), marker='o', label='Classical', color='red')
plt.plot(x, np.log10(quantum_train_loss), marker='o', label='Quantum', color='indigo')
plt.xlabel('Epoch')
plt.ylabel('Log of Loss')
plt.title('Log of Classical and Quantum LSTM Training Loss vs Epoch')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# calculate differences from paper results
def percent_error(expected, actual):
    dif = expected - actual
    return 100 * dif / expected

# values from john's training
classical_train_RMSE = 0.004553718874126597
classical_test_RMSE = 0.004869096496934205

quantum_train_RMSE = 0.014377863342364072
quantum_test_RMSE = 0.007396690897391019

# compare with paper results
print("classical train RMSE error: " + str(percent_error(0.0567, classical_train_RMSE)) + "%")
print("classical test RMSE error: " + str(percent_error(0.0693, classical_test_RMSE)) + "%")
print()
print("quantum train RMSE error: " + str(percent_error(0.0371, quantum_train_RMSE)) + "%")
print("quantum test RMSE error: " + str(percent_error(0.0602, quantum_test_RMSE)) + "%")
```
